### GRADE CALCULATOR
#### Program to determine letter grade for students enrolled in a class. 
#### This program is a general program to handle any number of classes, any number of students in each class, and any number of exams in each class. Happy Coding!


### Grading Comments: 

1. On line 29, why use a while loop (condition-controlled loop) when the user provided the number of exams that each student took in a class. Use for exam_number in range(1, num_of_exams+1,1): This way you do not need two statements to ask users for exam score
2. on line 36, Should use named constants instead of 0 and 100 because these limits may change in the future. 

```python
# Initialize all control variables and accumulators

continue_entry = 'Y'
class_number = 1 
A_LIMIT = 88.00
B_LIMIT = 78.00
C_LIMIT = 68.00
D_LIMIT = 60.00

# Outer while loop for Classes
while continue_entry == 'Y': 
    
    # Inputs
    num_of_students = int (input (f"Please enter the number of students in Class # {class_number}: "))
    num_of_exams = int (input (f"Please enter the number of exams in Class # {class_number}: "))

    # Inner loop for for each student and exam
    # PROCESS   
    for student in range (1,num_of_students+1):
        
        #Initialize counters for each loop and take input from User
        exam_number = 1
        total_score = 0
        average_score = 0   #MM Comments: why does average_score need to be zeroed out?
        
        student_name = input (f"Please enter full name (First , Last) of Student # {format(student)}: ")
        
        ## MM COMMENTS: Why use conditioned controlled loop when you the number of exams given in class!!
        while exam_number <= num_of_exams: 
            score = float (input (F"Please enter grade for Exam # {format(exam_number)} : "))
            
            #If loop to check validity of score
            
            ## MM COMMENTS _ Use named constants for upper and lower limits
            
            if score <0 or score >100:
                print ("Invalid data entry. Score must be between 0.00 and 100.00")
                score = float (input (F"Please enter grade for Exam # {format(exam_number)} : "))
            
            total_score += score                                     
            exam_number += 1
        
        average_score = total_score / num_of_exams
        
        ## Missing a check on Average Score being less than zero or greater than 100.0 and its corresponding error msg
        # Calculate grades based on average score and assign proper grade
        if average_score >= A_LIMIT :                                               
            letter_grade = 'A'                                                       
        elif average_score >= B_LIMIT:
            letter_grade = 'B'
        elif average_score >= C_LIMIT:
            letter_grade = 'C'
        elif average_score >= D_LIMIT:
            letter_grade = 'D'
        else:
            letter_grade = 'F'
    
    # OUTPUT
    
        print ('\n------------------Results-------------------\n')
        print ('Student Name\t\t\t: \t', student_name)
        print (F"Total of {format(num_of_exams)} exams\t\t: \t {format(total_score,'.2f')}")  
        print (F"Number of Exams - Class # {format (class_number)}\t: \t {num_of_exams}")
        print (F"Average Grade\t\t\t: \t {format(average_score,'.2f')}")
        print ('Letter Grade Assigned\t\t: \t', letter_grade)
        print ('\n--------------------------------------------\n')
     
        student += 1                                                        # Set counter for next student
    
    continue_entry = input("Do you wish to continue with next class? (Y or N):").upper()
    class_number += 1                                                       # Set counter for next Class

print ("Program Terminated by User. End of Program.")    
    
    
```
