```python
# A program to process departmental payroll for a particular department.
# This program will read payroll related data from an payroll_input CSV file,
# compute regular gross pay, overtime gross pay, total gross pay , tax to withhold based on hourly pay rate
# deductions based on total gross pay and net pay after subtracting tax and deductions from total gross pay.
# Wite the output to the console as well as to a CSV Formatted ShainanAgrawal_payroll_output file
# ----------------------------------------------------------------
# 1. Open the payroll_input.csv for reading in the payroll related data (read mode)
# 2. Open the ShainanAgrawal_payroll_output.csv for writing in the computed data (write mode)
# 3. Define Maximum weekly hours as 40. Any hour over 40 is overtime hour. Overtime pay is 1.5 times more.
# 4. Define overtime pay constant as 1.5.Define Tax rate constants based on hourly pay rate. 
# 5. Define deduction constants based on Grsoss pay.Set count entries=0 to count the entries amde in the csv file.
# 6. read all the record from the input file and assign to payroll_input_record.Strip new line character.
# 7. For each record, Split the values and assign them to correct variables.
# 8. set overtime hours and overtime pay as 0and then start checking for cpndition. - 
#      8.1 If weekly hours is less than maximum weekly hours, calculate regular gross pay.
#      8.2 Else calculate overtime hours and oevrtime pay.
#      8.3 Claculate gross pay as regular_gross_pay + overtime_gross_pay
# 9. Calculate Tax based on hourly pay rate as total_gross_pay * respective tax rate / 100
#      9.1 If hourly pay rate is below 12, tax rate is 10%
#      9.2 If hourly pay rate is 12 or below 20, tax rate is 15%. Else its 20%
# 10. Based on total gross pay, calculate deductions.
#      10.1 If gross pay is less than $800, deduction rate is 2 % else 5%
# 11. Calculate net_pay as total_gross_pay - tax_withheld - deduction_withheld
# 12. Concatenate results and assign to employee_record so each is separated by a comma
# 13.Incrcease value of count record by 1 in each loop
# 14.write the values to the write payroll_output.csv file.
# 15.Print the output to console and Print the number of entries made in the file using count variable.
# 16.Close input and output file. 
# --------------------------------------------------------------

# Initialization
payroll_input = open('/Users/cis_developer/CIS5357Exam/payroll_input.csv','r')
payroll_output = open ('/Users/cis_developer/CIS5357Exam/payroll_output.csv','w')

MAXIMUM_WEEKLY_HOURS = 40
OVERTIME_PAY_CONSTANT = 1.5
TAX_RATE_LOW = 10
TAX_RATE_MEDIUM = 15
TAX_RATE_HIGH = 20
HOURLY_PAY_RATE_LOW = 12
HOURLY_PAY_RATE_MEDIUM = 20
DEDUCTION_THRESHOLD = 800.00
DEDUCTION_RATE_LOW = 2
DEDUCTION_RATE_HIGH = 5


count_entries = 0

# inputs - read the record from the input file and split each record.
payroll_input_record = payroll_input.read().split('\n')

for record in payroll_input_record:

    #For each record, Split the values and assign them to correct variables.
    payroll_input_record_split = record.split(',')
    employee_name = payroll_input_record_split[0]
    weekly_hours_worked = float (payroll_input_record_split[1])
    regular_hourly_pay_rate = float (payroll_input_record_split[2])
   
    # Process
    overtime_hours = 0
    overtime_gross_pay = 0
    overtime_pay_rate = OVERTIME_PAY_CONSTANT * regular_hourly_pay_rate
    
    # Check if weekly hours worked is less than maximum weekly hours, If yes, calculate gross payement 
    if (weekly_hours_worked <= MAXIMUM_WEEKLY_HOURS):
        regular_hours_worked = weekly_hours_worked
        regular_gross_pay = weekly_hours_worked * regular_hourly_pay_rate
        
    # if weekly hours worked is more than maximum weekly hours, calculate overtime payment     
    else:
        regular_hours_worked = MAXIMUM_WEEKLY_HOURS
        regular_gross_pay = MAXIMUM_WEEKLY_HOURS * regular_hourly_pay_rate
        overtime_hours = weekly_hours_worked - MAXIMUM_WEEKLY_HOURS
        overtime_gross_pay = overtime_hours * overtime_pay_rate
    
    # Calcluate total gross pay based on overtime pay calculation
    total_gross_pay = regular_gross_pay + overtime_gross_pay
    
    # Check if hourly pay rate is low, medium or high and calculate tax based on respective tax rates.
    if regular_hourly_pay_rate < HOURLY_PAY_RATE_LOW:
        tax_withheld = (total_gross_pay * TAX_RATE_LOW)/ 100
    elif regular_hourly_pay_rate < HOURLY_PAY_RATE_MEDIUM:
        tax_withheld = (total_gross_pay * TAX_RATE_MEDIUM)/ 100
    else:
        tax_withheld = (total_gross_pay * TAX_RATE_HIGH)/ 100
    
    # Based on defined deduction threshold limit, calculate deduction to withhold as per deduction late.
    if total_gross_pay < DEDUCTION_THRESHOLD:
        deduction_withheld = (total_gross_pay * DEDUCTION_RATE_LOW) /100
    else:
        deduction_withheld = (total_gross_pay * DEDUCTION_RATE_HIGH) /100
    
    #Calculate net pay based on total gross pay, tax to withhold and deduction to withhold
    net_pay = total_gross_pay - tax_withheld - deduction_withheld
    
    #Process employee record to write to the output file in CSV format
    employee_record = (f"{employee_name},{regular_hours_worked},{regular_hourly_pay_rate},{regular_gross_pay},\
                       {overtime_hours},{format(overtime_pay_rate,'.2f')},\
                       {format(overtime_gross_pay,'.2f')},{format(total_gross_pay,'.2f')},\
                       {format(tax_withheld,'.2f')},{format(deduction_withheld,'.2f')},{format(net_pay,'.2f')}")
    count_entries += 1

    #Write to file
    payroll_output.write(employee_record +'\n')    
        
    # OUTPUT
    
    print ("\nEmployee’s Name:\t\t",employee_name)
    print (f"Regular Hours Worked :\t\t  {format(regular_hours_worked,'8.2f')}")
    print (f"Hourly Pay Rate :\t\t$ {format(regular_hourly_pay_rate,'8.2f')}")
    print (f"Regular Gross Pay :\t\t$ {format(regular_gross_pay,'8.2f')}")
    print (f"Overtime Hours Worked :\t\t  {format(overtime_hours,'8.2f')}")
    print (f"Overtime Pay Rate :\t\t$ {format(overtime_pay_rate,'8.2f')}")
    print (f"Overtime Gross Pay :\t\t$ {format(overtime_gross_pay,'8.2f')}")
    print (f"Total Gross Pay :\t\t$ {format(total_gross_pay,'8.2f')}")
    print (f"Taxes Withheld :\t\t$ {format(tax_withheld,'8.2f')}")
    print (f"Deductions Withheld :\t\t$ {format(deduction_withheld,'8.2f')}" )
    print (f"Net Pay :\t\t\t$ {format(net_pay,'8.2f')}")
    print ('\n--------------------------------------------')

    
# Finalization -- displays the number of entries made to the output file
print (f"\nCompleted Payroll Processing for {count_entries} employees.\n")
print (f"{count_entries} employee records were written to the output file. \n")
    
payroll_input.close()   # Close input file
payroll_output.close()  # Close output file
```
