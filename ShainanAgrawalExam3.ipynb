```python
def read_currency_data():
    
    import json
    import requests

    url_string = "https://api.exchangeratesapi.io/latest?base=USD"
    url_response = requests.get(url_string)
    exchange_rates = json.loads(url_response.text)
    exchange_rates_dict = exchange_rates['rates']
    
    return exchange_rates_dict
```

```python
def get_currency_exchange_rate(exchange_rates_dict):

    success = 0
    
    while success == 0:
        currency_for_conversion = input("\nWhat currency would you like to convert to (3-letter currency symbol)?:\t").upper()
        
        if currency_for_conversion in exchange_rates_dict:
            exchange_rate = exchange_rates_dict[currency_for_conversion]
            print(f"Exchange rate is available for {currency_for_conversion} currency.\nExchange rate is:{exchange_rate}")
            success=1
        else:
            print(f"No Exchange Rate is available for {currency_for_conversion}.\nKindly enter a symbol for which an exchange rate is available.")
            
    return currency_for_conversion, exchange_rate
    
```
